```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
df = yf.download('AAPL',
                 start='2010-01-01',
                 end='2021-01-01',
                 progress=False)
#Keep adjusted close prices only
df = df.loc[ :,['Adj Close']]
```

```python
#calculate simple and log returns
df['simple_rtrn'] = df['Adj Close'].pct_change()
df['log_rtrn'] = np.log(df['Adj Close']/df['Adj Close'].shift(1))
df = df.dropna()
df.head()
```
